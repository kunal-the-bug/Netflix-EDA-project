# Trends over Time.

```python
#Ensure 'release_year' is in numeric format
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
```

```python
# Drop rows with missing release_year
df = df.dropna(subset=['release_year'])
```

```python
#Group by release year and calculate the average duration
yearly_avg_duration = df.groupby('release_year')['duration'].mean()
```

```python
#Plot the trends over time
plt.figure(figsize=(12, 6))

plt.plot(yearly_avg_duration.index, yearly_avg_duration.values,color='brown')

plt.xlabel('Release Year')
plt.ylabel('Average Duration ( In minutes)')
plt.title('Average Movie/Episode Duration Over Time')
plt.grid(True)
plt.show()
```

# Content Evolution over Time

```python
content_by_year = df['release_year'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(content_by_year.index, content_by_year.values, marker='o', color='k')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.title('Evolution of Netflix Content Over Time')
plt.grid()
plt.show()
```
